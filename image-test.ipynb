```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("dataset/marah/1564075.jpg")

img_gray = cv2.imread('dataset/marah/1564075.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img)
plt.axis('off')
plt.show()

plt.imshow(img_gray)
plt.axis('off')
plt.show()

print("===== Original Images =====")
print(img.shape)
print(img.dtype)

print("===== Grayscale =====")
print(img_gray.shape)
print(img_gray.dtype)
```

```python
print("===== Nilai Matrix =====")
print(img)

print("===== Nilai Matrix (Grayscale) =====")
print(img_gray)
```

```python
plt.figure(figsize=(10, 8))
plt.hist(img_gray.ravel(), bins=256, range=[0, 256], color='gray')
plt.title('Histogram')
plt.xlabel('Pixel Density')
plt.ylabel('Number of pixels')
plt.grid(True)
plt.show()
```

```python
colors = ('r', 'g', 'b')
channel_ids = (0, 1, 2)

plt.figure(figsize=(10, 8))

for channel_id, color in zip(channel_ids, colors):
    plt.hist(
        img[:, :, channel_id].ravel(),
        bins=256,
        range=(0, 256),
        color=color,
        alpha=0.5,  # transparansi biar ga nutupin channel lain
        label=f'{color.upper()} channel'
    )
plt.title('Histogram RGB')
plt.xlabel('Pixel Density')
plt.ylabel('Number of Pixels')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd

# Simpan matrix warna (RGB)
img_rgb_flat = img.reshape(-1, img.shape[2])  # reshape supaya rapi kalo mau simpan RGB
df_rgb = pd.DataFrame(img_rgb_flat, columns=['R', 'G', 'B'])
df_rgb.to_csv('matrix_rgb.csv', index=False)

# Simpan matrix grayscale
df_gray = pd.DataFrame(img_gray)
df_gray.to_csv('matrix_gray.csv', index=False)
```
